## 🔑 Step 2: Configure Environment Secrecy (Gemini API Key)
The AI Coach and Food Analyzer require a Google Gemini API Key. You can get yours from [Google AI Studio](https://aistudio.google.com/).

Run the cell below to prompt interactively for your API Key and write it into a local `.env` configuration file safely.

```python
import os
from getpass import getpass

print("Please provide your Gemini API Key. It will be hidden as you type:")
api_key = getpass("GEMINI_API_KEY: ")

if api_key:
    # Write to local .env file
    with open(".env", "w") as f:
        f.write(f'GEMINI_API_KEY="{api_key}"\n')
        f.write('APP_URL="http://localhost:3000"\n')
        f.write('NODE_ENV="development"\n')
    print("\nConfig file .env generated successfully!")
else:
    print("\nWarning: No key entered. AI features will require you to add the GEMINI_API_KEY inside .env manually before starting.")
```
